**Data Cleaning :**  

1. Deleting the Particular rows and columns. ( Doing only missing values are less than 30 % )
2. Handle the missing values. (Replace the Min / Max / Mean / Median / Mode )

```python
# Find the missing values in the dataframe

df.isnull().sum()
```

### **1. Delete the missing values :**

```python
# Delete the missing values in the dataframe ( Only doing for missing values less than 30 % )

data_clean1 = df.dropna()
data_clean1
```

```python
# Check the clean dataframe after deleting the missing values

data_clean1.isnull().sum()
```

### **2. Handling the missing values :**  
* Delete the missing values in the Indipendent variable ( Target variable ) .
* Separate the values of categorical variables and numerical variables.
* Replace the value ( min / max / mean / median / mode ) in the numerical and categorical variables.

```python
# Deleting the missing values in the target variable

data_clean2 = df.dropna(subset = ['Loan_Status'])
data_clean2
```

```python
# Separate the categorical values in the dataframe & replace the mode value in the missing values

categorical = ['Gender', 'Area', 'Loan_Status']
data_clean2[categorical] = data_clean2[categorical].fillna(data_clean2.mode().iloc[0])
data_clean2[categorical]
```

```python
# Separate the numerical values in the dataframe & replace the mean value in the missing values

numerical = ['ApplicantIncome',	'CoapplicantIncome',	'LoanAmount']
data_clean2[numerical] = data_clean2[numerical].fillna(data_clean2.mean().iloc[0])
data_clean2[numerical]
```

```python
# Check the clean dataframe after deleting missing values

data_clean2.isnull().sum()
```
