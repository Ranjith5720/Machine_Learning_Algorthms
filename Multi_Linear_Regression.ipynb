### **Train & Test split :**


```python
# Separate the Independent variable

x = df.iloc[:,:-1]
x
```

```python
# Separate the Dependent varaiable

y = df.iloc[:,-1]
y
```

```python
# train , test split

from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size= 1/3, random_state= 13)
```

### **Model Building :**
**Multi - Linear Regression :**
* **y = m0 + m1 x1 + m2 x2 + . . . + mn xn + c**  
( x1 + x2 + x3 + . . . + xn ) - Independent variables  
( m1 + m2 + m3 + . . . + mn ) - Correlation co-efficient  
y - Dependent variable  
c - Constant  
m0 - Bias


```python
# Import LinearRegression

from sklearn.linear_model import LinearRegression
```

```python
mlr = LinearRegression()
```

```python
mlr.fit(x_train , y_train)
```

```python
# Model evaluation

y_predict = mlr.predict(x_test)
y_predict
```

```python
# Find difference for the test and predicted dataset

y_predict - y_test
```

```python
# Visualisation for the predicted data using scatter plot

plt.scatter(y_test,y_predict)
```
