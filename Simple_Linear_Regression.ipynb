**Train & Test Split :**

```python
# Saparate the Indepentend variable

x = df.iloc[:,:-1]
x
```

```python
# Separate the Depentend variable

y = df.iloc[:,-1]
y
```

```python
# Split the train & test data

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/3,random_state=7)
```

### **Model Building :**  
**Simple Linear Regression :**
* y = m x + c  
x - Independent variable  
y - Dependent variable  
m - Coefficient  
c - Constant

```python
# Import LinearRegression

from sklearn.linear_model import LinearRegression
```

```python
slr = LinearRegression()
```

```python
slr.fit(x_train, y_train)
```

```python
# Model Evaluation

y_predict = slr.predict(x_test)
y_predict
```

```python
# Visualisation of prdicted model

plt.scatter(y_test,y_predict)
```
