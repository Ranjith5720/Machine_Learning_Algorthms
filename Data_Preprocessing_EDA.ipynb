**Data Preprocesing :**  
Data preprocessing is a process of preparing the raw data and making it suitable for a machine learning model.

A real-world data generally contains noises, missing values, and maybe in an unusable format which cannot be directly used for machine learning models.

Data preprocessing is required tasks for cleaning the data and making it suitable for a machine learning model which also increases the accuracy and efficiency of a machine learning model.

**Stages of data preprocessing :**

1. Getting the dataset  
2. Importing libraries   
3. Importing datasets  
4. Find Exploratry Data Analysis ( EDA )
5. Finding Missing Data  
6. Encoding Categorical Data  
7. Splitting dataset into training and test set  
8. Feature scaling  
9. data normalization

**Dataframe Slicing :**

```python
# Separate the row values ( Positive indexing )

df.iloc[4]
```

```python
# Separate the row values ( Negative indexing )

df.iloc[-1]
```

```python
# Separate the particular value in the dataframe. 4 - row , 5 - columns

df.iloc[4,5]
```

```python
df.iloc[5,1:3]
```

```python
# Select defaultly first to last rows

df.iloc[:]
```

```python
# Select start(0) to end(n-1) rows

df.iloc[0:7]
```

```python
# Select specific index column(3rd) in the dataframe

df.iloc[:,3]
```

```python
# Select specific index column(5th) in the dataframe

df.iloc[0:7,5]
```

```python
# Select specific rows(start=0, end=n-1(6)) and columns(start=1, end=n-1(4)) in the dataframe

df.iloc[0:7,1:5]
```

```python
# Find the missing values in the dataframe

df.isnull().sum()
```
