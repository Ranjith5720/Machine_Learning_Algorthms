**Dataframe Slicing :**

```python
# Separate the row values ( Positive indexing )

df.iloc[4]
```

```python
# Separate the row values ( Negative indexing )

df.iloc[-1]
```

```python
# Separate the particular value in the dataframe. 4 - row , 5 - columns

df.iloc[4,5]
```

```python
df.iloc[5,1:3]
```

```python
# Select defaultly first to last rows

df.iloc[:]
```

```python
# Select start(0) to end(n-1) rows

df.iloc[0:7]
```

```python
# Select specific index column(3rd) in the dataframe

df.iloc[:,3]
```

```python
# Select specific index column(5th) in the dataframe

df.iloc[0:7,5]
```

```python
# Select specific rows(start=0, end=n-1(6)) and columns(start=1, end=n-1(4)) in the dataframe

df.iloc[0:7,1:5]
```
